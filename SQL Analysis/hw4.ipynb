## Step 3: Investigate the data

```python
SQLjoin = '''SELECT avg(rating), reviews.podcast_id, categories.category FROM reviews INNER JOIN categories on
            categories.podcast_id=reviews.podcast_id GROUP BY category ORDER BY avg(rating) DESC;'''
review_categories = pd.read_sql_query(SQLjoin, con).reset_index()
```
